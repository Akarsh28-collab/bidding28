```python
"""
--------------------------------------------------------------------
BIDDING WAR AI ‚Äì Competitive Artificial Agents (Short Version)
--------------------------------------------------------------------
Input:
User gives item prices separated by space.
Example: 100 120 80 200

Agents:
A = Aggressive bidder (bids high when item price is high)
B = Conservative bidder (bids small increments)

Output:
Step-by-step bidding history and winner summary
--------------------------------------------------------------------
"""

import random

# ---------------- USER INPUT ----------------
prices = list(map(int, input("Enter item base prices separated by space: ").split()))
n = len(prices)

cashA = cashB = 1000   # starting budget
scoreA = scoreB = 0
log = []

# ---------------- BIDDING ROUND LOOP ----------------
for i, price in enumerate(prices, start=1):

    # Agent A: aggressive (bids high for costly items)
    bidA = price + random.randint(15, 60)

    # Agent B: conservative (bids small increments)
    bidB = price + random.randint(5, 35)

    # If no budget left, forced low bids
    if cashA <= 0: bidA = 0
    if cashB <= 0: bidB = 0

    # Determining who wins this round
    if bidA > bidB and bidA <= cashA:
        cashA -= bidA
        scoreA += 1
        log.append(f"Round {i}: A wins (A={bidA}, B={bidB})")
    elif bidB > bidA and bidB <= cashB:
        cashB -= bidB
        scoreB += 1
        log.append(f"Round {i}: B wins (A={bidA}, B={bidB})")
    else:
        log.append(f"Round {i}: No one won due to budget issues (A={bidA}, B={bidB})")

# ---------------- FINAL OUTPUT ----------------
print("\n----- BIDDING HISTORY -----")
for l in log: print(l)

print("\n----- FINAL SCORECARD -----")
print("Items won by Agent A:", scoreA)
print("Items won by Agent B:", scoreB)
print("Remaining money A:", cashA)
print("Remaining money B:", cashB)

if scoreA > scoreB:
    print("üèÜ WINNER ‚Üí Agent A")
elif scoreB > scoreA:
    print("üèÜ WINNER ‚Üí Agent B")
else:
    print("ü§ù DRAW ‚Äì equal items won")
```
